# knn approach to finding traces with flipped polarity in seismic data

We use ML to classify the polarity $(\pm 1)$ of seismic traces. Each trace corresponds to a hydrophone which listens for sound wave reflecting from the geology in the subsurface of the Earth. The sound waves have denotations as the impulse sources which we refer to as "shots". Multiple arrays of receivers acquire the reflective waves coming from the geology in the subsurface. We call these arrays of receivers "channels". Sometimes the channel arrays have geophones which are recording data with a phase inversion of pi. That is, they record -f(t) when they should be recording f(t), where f(t) is the reflected wave. 

To process seismic data into a tomographic image, these flipped polarity traces need to be identified. The open source seismic data we utitlize in this project comes with channels that have this error. We use a nearest-neighbors algorithm to catalog them.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap


import pandas as pd, numpy as np
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import Normalizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
df= pd.read_csv('seismic_traces.txt',header=None,skiprows=[0])
```

Each channel contains 51 samples of data (amplitudes collected at each time sample). The data we are using has all channels concatenated into one series:

```python
df.shape
```

So we need to reorganize from a series into a dataframe where each channel populates its own column. We have different shots and channels, each sample has been labeled by these values as well as whether it is inverted or not. Label = 1 means no inversion (good), label = -1 is inversion (bad).

```python
df.head(10)
```

```python
df2 = pd.DataFrame(index=range(0,53))

for x in range(int(len(df)/2)): 
    if x % 52 == 0:
        df2[x] = pd.Series(df.iloc[x:x+53,0].reset_index()[0])
```

```python
def breakString(x):
    ls = x.split()
    return ls[len(ls)-1]

def pickSign(str):      # function to pick send gut=-1 to -1, gut=1 to 1
    if str[-2:-1]=='-':
        return -1
    else:
        return 1
```

```python
df2.head()
```

```python
df3 = df2.applymap(breakString).transpose()
df3['Polarity']= df3[0].apply(pickSign)

#We have the target column now, so we can drop the gut=+/-1 column.
#Write out dataframe

df4 = df3.drop(labels=[0,52],axis=1)
```

**Now we have the data in the shape that we want. Every row corresponds to a data point, with the entries being the samples of the recorded sound waves. The last attribute is the Polarity value of +/- 1.**

```python
df4.head(6)
```

### Building the nearest-neighbors model
We proceed to split the data for training/testing and build the model. We also split a validation set.

```python
X = df4.drop(labels='Polarity',axis=1)
y = df4['Polarity']*-1
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

We test to find the optimal k, the number of neighbors, via cross validation using cross_val_score. Since this is a classification problem, we test for accuracy.

```python
cv_scores = []
for n in range(1,14,2):
    knn = KNeighborsClassifier(n_neighbors=n,weights='uniform')
    scores = cross_val_score(knn, X_train, y_train, cv=3, scoring='accuracy')
    cv_scores.append(scores.mean())
```

```python
# Let's plot the misclassification error, call it msc
msc = [1 - x for x in cv_scores]
print('\n Accuracies are ', cv_scores)

n_vals = (list(range(1,14,2)))
optimalMSC = n_vals[msc.index(min(msc))]
print('\n Optimal k =', optimalMSC)
plt.figure(figsize=(7,7))
plt.plot(n_vals, msc)

plt.title('Error(k)')
plt.xlabel("Number of Neighbors k")
plt.ylabel("Misclassification Error")
plt.show()
```

We can look at few of the predictions:

```python
nrnb1 = KNeighborsClassifier(n_neighbors=1, weights='uniform').fit(X_train,y_train)
print('Score for k = 1 knn-model', nrnb1.score(X_test,y_test))
```

```python
y_predicted = nrnb1.predict(X_test)
```

```python
y_predicted[:20]
```

### Can we improve the model via some l2-scaling in preprocessing?

Answer: We get , but minimally so.

```python
transformer = Normalizer(norm='l2').fit(X_train)  
X_train_scaled = transformer.transform(X_train)
X_test_scaled = transformer.transform(X_test)
```

```python
nrnb1_scaled  = KNeighborsClassifier(n_neighbors=1, weights='uniform').fit(X_train_scaled,y_train)
print('Using l2-normalization on each observation gives an accuracy of the k = 1 model of', nrnb1_scaled.score(X_test_scaled,y_test))
```

We saw above that we could get an accuracy of 0.94 without using the normalizer.

## Can we use only two attributes?

Instead of using all 51 attributes for each channel, how close can we get to predicting polarity if we use the first two amplitude samples only.

The answer is, it depends. But the 51-dimensional case wins. For example, we will see that if we use X[[10,20]], that is, the 10th and 20th samples, then we don't even beat the most-frequent dummy classfier. Then we will see that if choose X[[1,2]], the 1st and 2nd samples, for our attributes, then we get a much improved accuracy but still lag behind the 51-dimensional model. 

```python
X2d_train_10and20,X2d_test_10and20,y2d_train_10and20,y2d_test_10and20 = train_test_split(X[[10,20]],y,random_state=0)
```

```python
cv_scores_2d = []
for n in range(3,50,2):
    knn = KNeighborsClassifier(n_neighbors=n,weights='uniform')
    scores = cross_val_score(knn, X2d_train_10and20, y2d_train_10and20, cv=5, scoring='accuracy')
    cv_scores_2d.append(scores.mean())
```
